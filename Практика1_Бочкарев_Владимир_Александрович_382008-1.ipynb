```python
# Пункт a

import xlrd
from collections import Counter

book_16_17 = xlrd.open_workbook('a/01_Образование_организации_01.xls')
book_15_18 = xlrd.open_workbook('a/01_Образование_организации_02.xls')
sheet_16_17 = book_16_17.sheet_by_name('Отчет')
sheet_15_18 = book_15_18.sheet_by_name('Отчет')

print(sheet_16_17.nrows)
print(sheet_15_18.nrows)
```

```python
# Пункт b

dct = {}
for row in range (4, sheet_15_18.nrows):
    key = sheet_15_18.cell_value(row,  0)
    if not('федеральный округ' in key):
        value = sheet_15_18.cell_value(row,  1)
        dct[key] = [value]
for row in range (4, sheet_16_17.nrows):
    key = sheet_16_17.cell_value(row,  0)
    if not('федеральный округ' in key):
        value16 = sheet_16_17.cell_value(row,  1)
        value17 = sheet_16_17.cell_value(row,  2)
        dct[key].append(value16)
        dct[key].append(value17)
for row in range (4, sheet_15_18.nrows):
    key = sheet_15_18.cell_value(row,  0)
    if not('федеральный округ' in key):
        value = sheet_15_18.cell_value(row,  2)
        dct[key].append(value)

# Исключим из словаря данные по субъектам, у которых нет данных за все 4 года
new_dct = new_dict = {k: [int(i) for i in v if i] for k, v in dct.items() if "" not in v}
dct = new_dct

for k,v in dct.items():
    print(k,":",v)
```

```python
# Пункт c

for k, v in dct.items():
    avg = sum(v)/len(v)
    dct[k].append(avg)

for k, v in dct.items():
    index = 0
    max = -1
    for i in range(len(v)-1):
        if (v[i]>max):
            max = v[i]
            index = i
    dct[k].append(2015+index)
    
for k,v in dct.items():
    print(k,":",v)
```

```python
# Пункты d, e

years = [value[-1] for value in dct.values()]
years_count = Counter(years)
sorted_dct = dict(sorted(dct.items(), key=lambda item: item[1][4]))

print(years_count,"\n")
for k,v in sorted_dct.items():
    print(k,":",v)
```
